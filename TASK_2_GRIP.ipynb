**IMPORTING THE REQUIRED LIBRARIES**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

**READING DATA FILE**

```python
iris=datasets.load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head() 
```

**CHECK FOR MISSING VALUES**

```python
iris_df.isnull().sum()
```

There is no missing values in the given dataset

**VISUALIZING THE DATA**

```python
iris_df.describe()
```

```python
iris_df.info()
```

```python
sns.pairplot(iris_df) #visualizing using pair plot
```

**FINDING THE OPTIMUM NUMBER OF CLUSTERS USING ELBOW METHOD**

```python
x=iris_df.iloc[:,[0,1,2,3]].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.show()
```

The optimum clusters is where the elbow occurs.

**APPLYING K MEANS CLUSTERING**

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

**VISUALIZING THE CLUSTERS**

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label="Iris-setoasa")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label="Iris-versicolour")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label="Iris-virginica")
plt.legend()
```

**PLOTTING WITH CENTROIDS OF THE CLUSTERS**

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label="Iris-setoasa")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label="Iris-setoasa")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label="Iris-setoasa")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="yellow",label='Centroids')
plt.legend()
```

So we have visually represented the clusters.

**CONCLUSION:**

The Optimum number of clusters is 3
